#1. 다음을 출력하는 프로그램을 작성하시오.

> **100** little **monkeys** jumping on the bed.

> One fell off and broke his head. 

> **Mama(or Daddy)** called the doctor and the doctor said, "No more monkeys jumping on the bed!"

> ...

> ... 

> **1** little **monkey** jumping on the bed. 

> One fell off and broke his head. 

> **Mama(or Daddy)** called the doctor and the doctor said, 

> "No more monkeys jumping on the bed!" 

(참고) Mama 혹은 Daddy를 랜돔하게 선택하게 하자.

```python
# 여기에 코드를 입력하세요.
import random
for num_monkey in range(100, 0, -1):
    if(num_monkey == 1):
        str_mingu = 'monkey'
    else:
        str_mingu = 'monkeys'
    list_parents = ["Mama", "Daddy"]
    selectedparents = random.choice(list_parents)
    list_mingu = ["his", "her"]
    selectedhisher = random.choice(list_mingu)
    print(str(num_monkey) + " little " + str_mingu + " jumping on the bed. \nOne fell off and broke " + selectedhisher + " head.\n" + selectedparents + " called the doctor and the doctor said, \n\"No more monkeys jumping on the bed!\"")
```

#2. 시작하는 연도와 끝나는 연도를 물어보는 프로그램을 작성 하고, 그 사이에 있는 모든 윤년을 출력하라. 

(시작하는 연도와 끝나는 연도가 윤년이면 그 해도 출력) 

* 윤년은 4로 나누어짐 (예: 1984, 2012)
* 100으로 나누어지면 윤년이 아님 (예: 1800, 1900) 
* 윤년은 400으로 나누어짐 (예: 1600, 2000) 

> Pick a starting year (like 1973 or something): 1973

> Now pick an ending year: 1977 

> Check it out... these years are leap years: 1976 

(참고) 윤년 리스트: http://mwultong.blogspot.com/2005/12/18002300-leap-year-list.html

```python
# 여기에 코드를 입력하세요.
year_start = int(input("Pick a starting year (like 1973 or something): "))
year_end = int(input("Now pick an ending year: "))
leap_year = 0
for i in range(year_start, year_end + 1):
    if((i%4 == 0) and (i%100 != 0)):
        leap_year = leap_year + 1
print("Check it out... these years are leap years: " + str(leap_year))
```

#3. 구구단을 출력하는 프로그램을 작성하고 결과를 txt 파일로 저장하시오.

```python
# 여기에 코드를 입력하세요.
file_mingu = open("result-w.txt","a")
for i in range(1, 10):
    for j in range(1, 10):
        file_mingu.write(str(i) + " X " + str(j) + " = " + str(j*i) + "\n")
    file_mingu.write("\n")
file_mingu.close()
```
